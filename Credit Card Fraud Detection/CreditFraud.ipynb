### Approaches to Fraud Detection 

Two primary approaches, Expert-Driven and Data-Driven approaches are sued for fraud detection. The Expert-Driven approach leverages the knowledge of fraud investigators to craft rules predicting the likelihood of fraud. These rules might be diverse. For example:
- Geographic Anomalies: If a card is used in multiple countries within a short timeframe, this could be flagged as suspicious. 
- Unusual High-Frequency Transactions: If there's an unusually high frequency of transactions in a short period, this could indicate card cloning.  
- Large Transactions from New Merchants: First-time transactions that are significantly higher than average could be suspect. Mismatched Location Data: If the physical location of a cardholder doesn’t match the transaction location, it may be fraudulent. 
- Uncharacteristic Purchases: If a transaction deviates from a cardholder's typical spending pattern, it might be fraudulent.   

These rules, based on patterns identified by experts, are straightforward to develop and understand, providing clear explanations for alerts. They capitalize on domain expertise but have limitations such as subjectivity, difficulty in discerning complex patterns, reliance on known fraud tactics, the necessity for ongoing human oversight, and a tendency to become outdated due to evolving fraud techniques. These limitations increase importance of data-driven approaches to fraud detection. 


### Data-Driven Approach for Fraud Detection 

The Data-Driven approach by the use of Machine Learning (ML) involves setting up a Fraud Detection System that learns from the data to identify probable fraud patterns. ML allows for the identification of complex fraudulent configurations, handling large volumes of data, and adapting to new fraud types and evolutionary changes.  In this project, we focus on data-driven ML techniques for fraud detection. Designing a Data-Driven Model based Fraud Detection System is complex, requiring decisions on features, strategies, algorithms, and model update frequency. This is particularly difficult given that fraud data is highly unbalanced, requiring a careful selection of machine learning algorithms and metrics. 

Please, also note we don't call for the use of only data-driven approach for fraud detection. We recognize that the ML-based data driven approach has drawbacks as well, including the need for substantial data samples and the "black box" nature of some models, which can be opaque to investigators. In real-world fraud detection systems (FDS), data-driven methods are typically employed to generate precise alerts and reduce false positives (FP). By contrast, expert-driven systems are predominantly used to ensure comprehensive fraud detection, thus minimizing false negatives (FN), albeit at the expense of a few false alerts. It's important to note that data-driven solutions should not be viewed as replacements for expert-based systems. Instead, they should be seen as complementary tools that enhance the accuracy of fraud detection. The best results often come from a hybrid approach, combining Data-Driven and Expert-Driven methods. The project explores a data-driven approach to fraud detection to complement expert-driven approach.   

We use the following techniques and metrics to develop Data-Driven Models for fraud detection:  
 
- Logistic Regression: We start with logistic regression classifier, since this is a good starting point for our binary classification problem. However, with highly imbalanced data, it might not perform optimally without further adjustments. Therefore, we move to ensemble learning approaches. 

- Ensemble learning: We first use random forest and then develop boosting algorithms such as Traditional Gradient Boosting Machines (GBM), AdaBoost, XGBoost, and LightGBM. Each of these algorithms have their own strengths and weaknesses. These ensemble methods can be more effective as they build multiple models and aggregate their results. They tend to perform well on imbalanced datasets.  

- Deep Learning Approaches: We develop both ANNs and CNNs. But note that CNNs are usually used for grid-style tasks and data such image processing. CNNs is used for practice purposes. 

- Resampling Techniques: While the use of appropriate method is important, it is equally important to use various sampling techniques to handle unbalanced data analysis. These include oversampling the minority class (fraud) or undersampling the majority class (legitimate). Both can help balance the dataset. Synthetic Minority Over-sampling Technique (SMOTE) is a popular method for oversampling.

- Evaluation Metrics:  The use of appropriate evaluation metrics for unbalanced data is important as well. For instance, accuracy is not a good metric for imbalanced datasets. Metrics like Precision, Recall, F1 Score, and AUC-PR (Area Under the Precision-Recall Curve) are more informative.

## <center> Preprocessiong 

In this section, we conduct preprocessing, making data ready for modeling. You can read below what pre-processing entails. We are going to apply only very few of these as our dataset has previously cleaned and used for previous research. 



Preprocessing involves cleaning and transforming our raw data into a format suitable for analysis and modeling. Some key steps in preprocessing include: (Please, note that these might overlap.)  

##### Data Cleaning: Handling Missing Values: 
   - fill in or remove missing data. You can check this pandas documentation: https://pandas.pydata.org/docs/user_guide/missing_data.html
   - remove duplicates
   - convert data to appropriate types. For instance, convert strings to numeric type, parse out dates 
   - address outliers in the data if there are any 
    

#### Feature Engineering:
   - Create new features, obtaining new columns from existing data.
   - For instance, encoding categorical data


##### Feature Transformation
1. normalization: this process involves scaling individual samples to have unit norm. This technique is useful when we have a dataset with varying scales and we want to equalize these scales without distorting differences in the ranges of values. A common approach is to scale the values between 0 and 1 or -1 and 1.
2. standardization: this involves rescaling the feature to have a mean of 0 and a standard deviation of 1. Unlike normalization, standardization does not bound values to a specific range, which might be useful for certain algorithms that assume data is centered around zero. It's useful in cases where the data follows a Gaussian distribution. In statistics, standardization is often referred to as computing the Z-score for each value or data point. 
3. log transformation: this is useful for handling skewed data. After transformation, data can resemble a normal distribution. Log transformation can help to stabilize data variance.
4. square root transformation: like log transformation, the square root can help in reducing skewness. It’s less strong than the log transformation.
5. more transformations: we can also do transformations like Box-Cox transformation, reciprocal transformation contingent on the specific requirements of the data distribution and the analytical model being used. 
6. do I apply transformation to the features or target variable? for **classification tasks**, transformations are typically applied only to the features (X columns). In classification tasks, the target variable (y) is categorical, so transformations like standardization or normalization are not applicable. For **regression tasks**, we can apply transformations to both the features and the target variable, especially if the target variable is skewed.



##### Dimensionality Reduction:
   - apply PCA or other techniques to reduce the number of features.

##### Train-Test Split: 
   - divide your dataset into separate sets for training and testing. Remember, a model is useful when it can predict unseen data. 




#### Which Transformation Method for Which ML Algorithm? 

1. Linear/logistic Regression:
    - Standardization is often useful because these models assume that all features are centered around zero and have variance in the same order. Features with larger scales can disproportionately influence the model.
    - Normalization can be used but not typically necessary unless features have different scales. 

2. Support Vector Machines (SVM), k-Nearest Neighbors (k-NN), and Neural Networks. 
    - Standardization or normalization is very important because these models are sensitive to the scale of input data. 
    - Log/Square Root Transformation is useful if features are highly skewed.

3. Decision Trees and Random Forest:
    - these models don't require standardization/normalization becuase they are not sensitive to the scale of the input features. Generally, scaling comes handy in algorithms, which are distance based, requiring Euclidean Distance. As name suggests, Random Forest is a tree-based ML algorithm. Therefore, it does not need feature scaling.
    - transformation for skewed data might be useful if it improves the interpretability or the distribution of the features.

4. Principal Component Analysis (PCA):
    - Standardization is crucial, as PCA is affected by the scale of the features.
    
5. Gradient Boosting Machines (like XGBoost, LightGBM):
    - like random forest, standardization or normalization is not necessary because these models are not sensitive to the scale of the features. generally, ensemble learning methods (other than Gradient Boosting) like bagging or voting classifiers do not inherently require feature scaling.  But if they include models that are sensitive to feature scale, scaling might be necessary.
    - transformation for skewed data can be applied if it helps the model to better understand the relationships in the data.

6. Clustering Algorithms (like K-Means, Hierarchical Clustering):
    - standardization or normalization is very important because these algorithms use distance measures that are sensitive to the scale of the data.

7. Time Series Analysis (like ARIMA, SARIMA):
    - log or box-Cox transformation is often used to stabilize variance and make series more stationary, which is an essential assumption in time series analysis.


9. Deep Learning/Neural Networks: 
    - consider using batch normalization, which normalizes the inputs of each layer. It helps in faster convergence and can also regularize the model to some extent.

10. Dimensionality Reduction Techniques (other than PCA): 
    - techniques like t-SNE and UMAP also benefit from feature scaling, as they are based on distance metrics.

11. Regularized Regression Models (Lasso, Ridge): 
    - standardization is crucial because regularization penalizes the magnitude of coefficients. unscaled features can lead to biased penalties.

12. How do I decide?
    - depends on the requirements and assumptions of the algorithm 
    - distribution and scale of features
    - nature of target variable, especially in regression tasks.
    - validate the effectiveness of a transformation on model performance through experimentation and cross-validation.
