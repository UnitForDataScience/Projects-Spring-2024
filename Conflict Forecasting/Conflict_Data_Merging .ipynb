# <center> Data Merging

```python
# necessary libraries 
import pandas as pd
```

```python
# import datasets: navco  
navco = pd.read_excel("NAVCO2-1_ForPublication.xlsx")
navco
```

```python
# import datasets: Quality of Government(standard time series version) 
qog= pd.read_csv("qog_std_ts_jan24.csv")
qog
```

```python
# create (or rename current ones) new variables to match intersection variables 
navco['cow'] = navco["loc_cow"]
qog['cow'] = qog["ccodecow"]
```

```python
# subset variables for merging: navco
navco1=navco[['cow', 'year', 'prim_meth']]
navco1
```

```python
# subset variables for merging: qog

qog1=qog[['cow', 'year',"pwt_hci", "pwt_pop", "pwt_rgdp", "vdem_gender", "wdi_acel", "wdi_afp", 
    "wdi_afpt", "wdi_agedr", "wdi_araland", "wdi_armexp", "wdi_belmedinc", 
    "wdi_empprmne", "wdi_empprne", "wdi_expeduge", "wdi_expmil", "wdi_expmilge", 
    "wdi_gdpcapcon2015", "wdi_gdpcapcur", "wdi_gdpgr", "wdi_gerp", "wdi_gerpf", 
    "wdi_gini", "wdi_gniatlcur", "wdi_incsh10h", "wdi_incsh10l", "wdi_inflation", 
    "wdi_internet", "wdi_litrad", "wdi_litradf", "wdi_migration", "wdi_pop", 
    "p_polity2", "ht_regtype", "ht_regtype1", "al_ethnic2000", "al_language2000", 
    "al_religion2000"]]
qog1
```

```python
# inner join by country and year
inner= pd.merge(qog1, navco1, on=['cow', 'year'], how='inner')
inner.shape
inner
```

```python
# left join
left= pd.merge(qog1, navco1, on=['cow', 'year'], how='left')
left.shape
```

```python
# why doesn't left join dataset have 15243 rows?
# because more compaign events happened in the same country and year.

left[["cow", "year"]].duplicated().sum()
```

```python
# right join
right= pd.merge(qog1, navco1, on=['cow', 'year'], how='right')
right.shape
```

```python
# outer join 
outer= pd.merge(qog1, navco1, on=['cow', 'year'], how='outer')
print(outer.shape)
outer
```
